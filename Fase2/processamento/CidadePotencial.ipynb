```python
import pandas as pd
import numpy as np
```

```python
df_cidades_pot = pd.read_csv('../dados/brutos/codigos_municipios_regioes_sp.csv', sep=';', encoding='latin1')
df_cidades_pot
```

```python
df_cidades_pot.dropna()
```

```python
df_cidades_pot = df_cidades_pot.dropna()
```

```python
# DDL
"""
CREATE TABLE dw_score.DimCidadePotencial (
    KeyCidadePotencial INT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
    NomeCidadePotencial VARCHAR(255) NOT NULL,
    UFCidadePotencial CHAR(2) NOT NULL,
    CodigoIBGE CHAR(7) NOT NULL
);
"""

df_cidades_pot = df_cidades_pot.rename(columns={
    'municipio': 'NomeCidadePotencial',
    'cod_ibge': 'CodigoIBGE'
})
df_cidades_pot
```

```python
df_cidades_pot_to_save = pd.DataFrame({
    'NomeCidadePotencial': df_cidades_pot['NomeCidadePotencial'],
    'UFCidadePotencial': 'SP',
    'CodigoIBGE': df_cidades_pot['CodigoIBGE']
})
df_cidades_pot_to_save
```

```python
df_cidades_pot_to_save['NomeCidadePotencial'] = df_cidades_pot_to_save['NomeCidadePotencial'].astype(str)
df_cidades_pot_to_save['UFCidadePotencial'] = df_cidades_pot_to_save['UFCidadePotencial'].astype(str)
df_cidades_pot_to_save['CodigoIBGE'] = df_cidades_pot_to_save['CodigoIBGE'].astype(str)

df_cidades_pot_to_save.to_csv('../dados/processados/cidades_sp_processadas.csv', sep=',', encoding='UTF-8', columns=['NomeCidadePotencial', 'UFCidadePotencial', 'CodigoIBGE'], index=False)
```
