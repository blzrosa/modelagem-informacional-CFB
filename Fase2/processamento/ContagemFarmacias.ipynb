```python
import pandas as pd
import numpy as np
```

```python
df_farmacias = pd.read_excel('../dados/brutos/farmacias_sp.xlsx')
df_farmacias
```

```python
df_farmacias.drop(0)
```

```python
df_farmacias = df_farmacias.drop(0)
df_farmacias = df_farmacias[df_farmacias['Código IBGE'] != 'Totais']
df_farmacias
```

```python
df_farmacias.loc[:, 'Estabelecimentos (Matriz + Filial)'] = df_farmacias['Estabelecimentos (Matriz + Filial)'].astype(int)
df_farmacias
```

```python
df_farmacias.rename(columns= {
                        'Estabelecimentos (Matriz + Filial)': 'QtdFarmacias',
                        'MUNICIPIO_DESCRICAO': 'NomeCidadeContagem',
                        'UF': 'UFContagem',
                        'Código IBGE':  'CodigoIBGE'
                    }, inplace=True)

"""DDL:
CREATE TABLE dw_score.DimContagemFarmacias (
    KeyContagemFarmacias INT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
    QtdFarmacias INT CHECK(QtdFarmacias >= 0),
    NomeCidadeContagem VARCHAR(255) NOT NULL,
    UFContagem CHAR(2) NOT NULL,
    CodigoIBGE CHAR(7) NOT NULL
);
"""

df_farmacias
```

```python
"""DDL:
CREATE TABLE dw_score.DimContagemFarmacias (
    KeyContagemFarmacias INT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
    QtdFarmacias INT CHECK(QtdFarmacias >= 0),
    NomeCidadeContagem VARCHAR(255) NOT NULL,
    UFContagem CHAR(2) NOT NULL,
    CodigoIBGE CHAR(7) NOT NULL
);
"""

df_farmacias_to_save = df_farmacias[['QtdFarmacias', 'NomeCidadeContagem', 'UFContagem', 'CodigoIBGE']]
df_farmacias_to_save
```

```python
df_farmacias_to_save.sort_values(by='QtdFarmacias', inplace=True, ascending=False)
df_farmacias_to_save
```

```python
df_farmacias_to_save['QtdFarmacias'] = df_farmacias_to_save['QtdFarmacias'].astype(int)
df_farmacias_to_save['NomeCidadeContagem'] = df_farmacias_to_save['NomeCidadeContagem'].astype(str)
df_farmacias_to_save['UFContagem'] = df_farmacias_to_save['UFContagem'].astype(str)
df_farmacias_to_save['CodigoIBGE'] = df_farmacias_to_save['CodigoIBGE'].astype(str)

df_farmacias_to_save.to_csv('../dados/processados/farmacias_sp_processadas.csv', sep=',', encoding='UTF-8', columns=['QtdFarmacias', 'NomeCidadeContagem', 'UFContagem', 'CodigoIBGE'], index=False)
```
