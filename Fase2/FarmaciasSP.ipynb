```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('farmacias_sp.xlsx')
```

```python
df
```

```python
farmacias = df[df['Código IBGE'] == 'Totais']
farmacias
```

```python
df_farmacias = farmacias.drop(['Código IBGE'], axis=1).rename(columns={'Estabelecimentos (Matriz + Filial)': 'COUNT', 'MUNICIPIO_DESCRICAO': 'MUNICIPIO'}).sort_values(by='COUNT', ascending=False)
df_farmacias
```

```python
df_farmacias.loc[df_farmacias['MUNICIPIO'] == "São Paulo (território)", 'MUNICIPIO'] = "São Paulo"
df_farmacias.loc[df_farmacias['MUNICIPIO'] == "Embu", 'MUNICIPIO'] = "Embu das Artes"
df_farmacias['MUNICIPIO'] = df_farmacias['MUNICIPIO'].str.replace("´", "'", regex=False)
df_farmacias['MUNICIPIO'] = df_farmacias['MUNICIPIO'].str.replace("-", " ", regex=False)
```

```python
df_farmacias.to_csv('FarmaciasSPCount.csv', index=False)
```
