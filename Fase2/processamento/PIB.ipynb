```python
import pandas as pd
import numpy as np
```

```python
df_pib = pd.read_excel('../dados/brutos/pib_por_setor_sp.xlsx', header=None)
df_pib
```

```python
setor_anos = df_pib.iloc[0, 2:]
np.unique(setor_anos)
```

```python
anos_por_setor = df_pib.iloc[1, 2:]
colunas_mapeadas = []
for setor, ano in zip(setor_anos, anos_por_setor):
    if pd.isna(setor) or pd.isna(ano):
        colunas_mapeadas.append(('Desconhecido', 'Desconhecido'))
    else:
        setor = str(setor).strip()
        ano = str(int(float(ano)))
        colunas_mapeadas.append((setor, ano))
df_dados = df_pib.iloc[2:, :].copy()
df_dados
```

```python
df_dados = df_dados.reset_index(drop=True).drop([0, 1])
df_dados
```

```python
novos_nomes = ['UF', 'Municipio'] + [f"{s}_{a}" for s, a in colunas_mapeadas]
df_dados.columns = novos_nomes
```

```python
df_dados.reset_index(drop=True, inplace=True)
df_dados.replace('-', np.nan, inplace=True)
pib_cols = [col for col in df_dados.columns if '_' in col and col.split('_')[0] in [
    'Administração Pública', 'Agropecuária', 'Comércio e Serviços', 'Indústria'
]]
df_long = pd.melt(
    df_dados,
    id_vars=['UF', 'Municipio'],
    value_vars=pib_cols,
    var_name='Setor_Ano',
    value_name='ValorPIB'
)
df_long
```

```python
# DDL
"""
CREATE TABLE dw_score.DimPIB (
    KeyPIB INT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
    ValorPIB INT,
    AnoPIB INT NOT NULL,
    SetorPIB VARCHAR(32) NOT NULL,
    NomeCidadePIB VARCHAR(255) NOT NULL,
    UFPIB CHAR(2) NOT NULL,
    CodigoIBGE CHAR(7) NOT NULL
)
"""
df_long[['SetorPIB', 'AnoPIB']] = df_long['Setor_Ano'].str.split('_', expand=True)
df_long['ValorPIB'] = df_long['ValorPIB'].astype(float).astype('Int64')
df_long['UFPIB'] = df_long['UF'].str.strip()
df_long['NomeCidadePIB'] = df_long['Municipio'].str.strip()
df_long
```

```python
len(np.unique(df_long['NomeCidadePIB']))
```

```python
df_cidades = pd.read_csv('../dados/processados/cidades_sp_processadas.csv')
df_cidades
```

```python
df_cidades['NomeCidadePotencial'] = df_cidades['NomeCidadePotencial'].str.upper()
df_cidades
```

```python
map_pib = set(np.unique(df_long['NomeCidadePIB']))
map_cidades = set(np.unique(df_cidades['NomeCidadePotencial']))
print('Nomes PIB')
for cidade in map_pib:
    if cidade not in map_cidades:
        print(cidade)
print('#'*20)
print('Nomes cidades')
for cidade in map_cidades:
    if cidade not in map_pib:
        print(cidade)
```

```python
names_to_map = {
    'BIRITIBA-MIRIM' : 'BIRITIBA MIRIM',
    'FLORÍNIA': 'FLORÍNEA',
    'SÃO LUÍZ DO PARAITINGA': 'SÃO LUIZ DO PARAITINGA',
    'ITAÓCA': 'ITAOCA',
}

df_long['NomeCidadePIB'] = df_long['NomeCidadePIB'].map(names_to_map).fillna(df_long['NomeCidadePIB'])
```

```python
df_joined = pd.merge(df_long, df_cidades, how='inner', left_on='NomeCidadePIB', right_on='NomeCidadePotencial')
df_joined
```

```python
len(np.unique(df_joined['NomeCidadePIB']))
```

```python
# DDL
"""
CREATE TABLE dw_score.DimPIB (
    KeyPIB INT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
    ValorPIB INT,
    AnoPIB INT NOT NULL,
    SetorPIB VARCHAR(32) NOT NULL,
    NomeCidadePIB VARCHAR(255) NOT NULL,
    UFPIB CHAR(2) NOT NULL,
    CodigoIBGE CHAR(7) NOT NULL
)
"""

df_to_save = df_joined[['ValorPIB', 'SetorPIB', 'AnoPIB', 'NomeCidadePIB', 'UFPIB', 'CodigoIBGE']].copy()
df_to_save
```

```python
df_to_save['ValorPIB'] = df_to_save['ValorPIB'].astype('Int64')
df_to_save['SetorPIB'] = df_to_save['SetorPIB'].astype(str)
df_to_save['AnoPIB'] = df_to_save['AnoPIB'].astype(int)
df_to_save['NomeCidadePIB'] = df_to_save['NomeCidadePIB'].astype(str)
df_to_save['UFPIB'] = df_to_save['UFPIB'].astype(str)
df_to_save['CodigoIBGE'] = df_to_save['CodigoIBGE'].astype(str)

df_to_save.to_csv('../dados/processados/pib_sp_processado_por_setor.csv', sep=',', encoding='UTF-8', columns=['ValorPIB', 'SetorPIB', 'AnoPIB', 'NomeCidadePIB', 'UFPIB', 'CodigoIBGE'], index=False)
```
