```python
import pandas as pd
import numpy as np
```

```python
df_pop = pd.read_csv('../dados/brutos/censo2022_sexo_idade.csv', sep=';', encoding='latin1')
df_pop
```

```python
df_pop = df_pop[df_pop['ano'] == 2022]
df_pop
```

```python
# DDL
"""
CREATE TABLE dw_score.DimPopulacao (
    KeyPopulacao INT GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
    ContagemPop INT NOT NULL,
    FaixaEtariaPop VARCHAR(32) NOT NULL,
    SexoPop CHAR(1) NOT NULL,
    NomeCidadePopulacao VARCHAR(255) NOT NULL,
    UFPopulacao CHAR(2) NOT NULL,
    CodigoIBGE CHAR(7) NOT NULL
);
"""

df_pop = df_pop.rename(columns={
    'cod_ibge': 'CodigoIBGE',
    'municipio': 'NomeCidadePopulacao',
    'sexo': 'SexoPop',
    'idade': 'FaixaEtariaPop',
    'populacao': 'ContagemPop'
})
df_pop
```

```python
print(np.unique(df_pop['SexoPop']))
```

```python
df_pop_to_save = df_pop[['ContagemPop', 'FaixaEtariaPop', 'SexoPop', 'NomeCidadePopulacao']].copy()
df_pop_to_save['UFPopulacao'] = 'SP'
df_pop_to_save['CodigoIBGE'] = df_pop['CodigoIBGE']
df_pop_to_save['SexoPop'] = df_pop_to_save['SexoPop'].str[0]
df_pop_to_save
```

```python
faixa_map = {
    '00 a 04': '00-04',
    '05 a 09': '05-09',
    '10 a 14': '10-14',
    '15 a 19': '15-19',
    '20 a 24': '20-24',
    '25 a 29': '25-29',
    '30 a 34': '30-34',
    '35 a 39': '35-39',
    '40 a 44': '40-44',
    '45 a 49': '45-49',
    '50 a 54': '50-54',
    '55 a 59': '55-59',
    '60 a 64': '60-64',
    '65 a 69': '65-69',
    '70 a 74': '70-74',
    '75 a 79': '75-79',
    '80 a 84': '80-84',
    '85 a 89': '85-89',
    '90 e +': '90+'
}

df_pop_to_save['FaixaEtariaPop'] = df_pop_to_save['FaixaEtariaPop'].map(faixa_map)
df_pop_to_save
```

```python
df_pop_to_save['ContagemPop'] = df_pop_to_save['ContagemPop'].astype(int)
df_pop_to_save['FaixaEtariaPop'] = df_pop_to_save['FaixaEtariaPop'].astype(str)
df_pop_to_save['SexoPop'] = df_pop_to_save['SexoPop'].astype(str)
df_pop_to_save['NomeCidadePopulacao'] = df_pop_to_save['NomeCidadePopulacao'].astype(str)
df_pop_to_save['UFPopulacao'] = df_pop_to_save['UFPopulacao'].astype(str)
df_pop_to_save['CodigoIBGE'] = df_pop_to_save['CodigoIBGE'].astype(str)

df_pop_to_save.to_csv('../dados/processados/populacao_sp_processada.csv', sep=',', encoding='UTF-8', columns=['ContagemPop', 'FaixaEtariaPop', 'SexoPop', 'NomeCidadePopulacao', 'UFPopulacao', 'CodigoIBGE'], index=False)
```
